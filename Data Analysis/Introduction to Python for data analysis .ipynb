# Get The Data

```python
%%time
%whos
import pandas as pd
# Get the data
poll_url = 'https://projects.fivethirtyeight.com/general-model/president_general_polls_2016.csv'
%time polls = pd.read_csv(poll_url)
# head() only returns the first 5 rows of the data 
# polls.head()
polls.sort_values('startdate').head()
# polls.sort_values('startdate')
# polls.sort_values('startdate').head()
```

```python
df_data = [[1900, '1-4 Years', 1983.8], [1901, '1-4 Years', 1695.0]]
df_columns = ['Year', 'Age Group', 'Death Rate']
mortality_df = pd.DataFrame(
    data =df_data,
    columns = df_columns
)
mortality_df
```

```python
# save and restore a DataFrame
# to_pickle(), to_cvs(), to_excel()
mortality_df.to_pickle('mortality_data.pkl')
mortality_df.to_excel('mortality_data.xlsx')
# read from the saved data
mortality_data = pd.read_pickle('mortality_data.pkl')
mortality_data
```

```python
data = pd.read_csv('mortality.csv')
# data.head()
# data.tail()
with pd.option_context(
        'display.max_rows',6,
        'display.max_columns', 3):
    display(data)
```

```python
# attributes of a dataframe
print("index: ", data.index)
print('columns: ', data.columns)
print('size', data.size)
print('shape', data.shape)
print('values', data.values)
data.columns = data.columns.str.replace(' ', '')
print('columns: ', data.columns)
```
